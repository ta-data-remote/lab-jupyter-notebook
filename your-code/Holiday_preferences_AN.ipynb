```python
def holiday_preferences_according_by_season(season):
    from IPython.display import YouTubeVideo
    
          
    if season == "winter":
        print('It\'s ski time, let\'s go to Borovets')

        video = YouTubeVideo("zUaY3K5GMGM", height=500, width=900, start = 27)

        display(video)
        
    elif season == "summer":
        print('It\'s beach time, let\'s go to Denia')

        video = YouTubeVideo("G01yjIjLRZw", height=500, width=900, start = 21)
        
        display(video)
        
    elif season == "spring" or season == 'autumn':
        print('no holiday my friend, wake up it\'s time to go to work')

    else: print('Wrong input - Try again')
            
        

season = input('Write winter or spring or summer or autumn: ')


holiday_preferences_according_by_season(season)
```
